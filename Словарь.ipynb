```python
# путь к файлу
path_kr = '/content/gdrive/My Drive/COLAB_Andemir/homeworks/dict_kab_ru.txt' 

# исходный файл
with open(path_kr, "w") as f:
  text = '''бостей - платье
вагъуэ - звезда
дзэ - войско, зуб
дыгъэ - солнце
кхъужь - груша
лъакъуэ - нога
напэ - лицо
плъыжь - красный
унэ - дом
шэ - вези, молоко, пуля'''

  f.write(text)

f.close()
```

```python
# читаем файл
with open(path_kr, 'r') as f:
  kab_ru = f.read().split('\n')
f.close()  
```

```python
d = {}

# новый словарь
for i in kab_ru:
  words = i.strip().split(' - ')
  kab = words[0]
  ru = words[1].split(', ')
  for key in ru:
    if key in d:
      d[key].append(kab)
    else:
      d[key] = [kab]

# сортировка значений словаря
for key in d:
  d[key].sort()         
```

```python
path_rk = '/content/gdrive/My Drive/COLAB_Andemir/homeworks/dict_ru_kab.txt' 
```

```python
g = open(path_rk, 'w')
for kab in sorted(d):
  g.write(kab + ' - ' + ', '.join(d[kab]) + '\n')
g.close()  
```
